# Data Pre-processing

```python
y = data1[['Disease']]
```

```python
data1["Symptoms"] = 0
```

```python
rows = data1.shape[0]
for i in range(rows):
    values = data1.iloc[i].values
    values = values.tolist()
    if 0 in values:
        data1["Symptoms"][i] = values[1:values.index(0)]
    else:
        data1["Symptoms"][i] = values[1:]
```

```python
data1.head()
```

```python
columns = data1[['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4',
       'Symptom_5', 'Symptom_6', 'Symptom_7', 'Symptom_8', 'Symptom_9',
       'Symptom_10', 'Symptom_11', 'Symptom_12', 'Symptom_13', 'Symptom_14',
       'Symptom_15', 'Symptom_16', 'Symptom_17']].values.ravel()
```

```python
columns
```

```python
symptom_ = pd.unique(columns)
symptom_ = symptom_.tolist()
symptom_ = [i for i in symptom_ if str(i) != "nan"]
```

```python
X = pd.DataFrame(columns = symptom_,index = data1.index)
X
```

```python
X["Symptoms"] = data1["Symptoms"]
```

```python
for i in symptom_:
    X[i] = X.apply(lambda x:1 if i in x.Symptoms else 0, axis=1)
```

```python
X["Disease"] = data1["Disease"]
X = X.drop("Symptoms",axis=1)
```

```python
X.head()
```

```python
data1['Disease'].unique()
```

```python
dict1 = {}
for i, value in enumerate(data1['Disease'].unique()):
  dict1[value] = i
```

```python
dict1
```

```python
for i in range(len(X['Disease'])):
  X['Disease'][i] = dict1[X['Disease'][i]]
```

```python
X.head()
```

# Training Model

```python
from sklearn.model_selection import train_test_split,cross_val_score
train, test = train_test_split(X,test_size=0.2)
X_train = train.drop("Disease",axis=1)
y_train = train["Disease"].copy().astype('int')
X_test = test.drop("Disease",axis=1)
y_test = test["Disease"].copy().astype('int')
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier()
random_forest.fit(X_train,y_train)
```

# Evaluasi Model

```python
y_pred = random_forest.predict(X_test)
```

```python
cross_val_score(random_forest,X_train,y_train,cv=10).mean()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
